```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.scatterplot(x='age', y='fare',
                data=titanic_df)
```

```python
sns.scatterplot(x='age', y='fare',
                data=titanic_df, hue='alive',style='class')
plt.title('Titanic RIP Case')
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age' , y = 'fare',
                data = titanic_df , hue = 'alive',
                style = 'class')
sns.lineplot(x = 'age', y = 'fare' , data = titanic_df, color = 'green')
plt.title('Titanic Data Analysis')
```

```python
sns.barplot(x='class', y='fare', data=titanic_df)
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex')
     
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'inferno',
            estimator = np.max)
```

```python
sns.barplot(y = 'class' , x = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'inferno',
            estimator = np.max,
            orient='h')
```

```python
sns.barplot(x = 'class' , y = 'fare' , data = titanic_df ,
            hue = 'sex',
            palette = 'inferno',
            ci=100,
            errcolor = '#9289db',
            errwidth = 20)
```
